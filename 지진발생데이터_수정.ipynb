#### 2016년~2020년 데이터 확인 

```python
occur_data['발생시각'].isnull().sum() # 결측치 확인
```

```python
occur_data['발생시각'] = pd.to_datetime(occur_data['발생시각'])
```

```python
occur_data['발생년도'] = occur_data['발생시각'].dt.year
```

```python
year_list = [] # 년도만 분리해서 데이터프레임 생성 
for year in range(2016, 2021):
    data_year = occur_data[occur_data['발생년도'] == year]
    year_list.append(data_year)
```

```python
occur_20162020 = pd.concat(year_list) # 해당 데이터 프레임 병합
```

```python
occur_20162020
```

총 637 건

```python
occur_20162020[occur_20162020['규모'] > 2.0]
```

```python
occur_20162020[occur_20162020['규모'] > 2.5]
```

2.0 이상은 545건, 2.5 이상은 161건 발생.
