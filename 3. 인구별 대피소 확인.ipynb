## 1. population에서 대피소total 에 있는 시군구명만 뽑아내기
- 시도+시군구명의 칼럼을 생성.
- 시군구명의 unique랑 시도+시군구명의 unique의 값이 다른 것을 확인.
- earth와 population의 시도_시군구명 동일화 처리.
- 병합 후 earth의 기존 row값 안 변한 것을 확인.

```python
earth['시도명'].unique()
```

```python
# earth 와 population 도시이름 같은지 확인
population['도시이름'].unique()
```

```python
#population을 earth랑 같게
population.loc[population['시군구'] == '포항시 북구','시군구'] = '포항시북구'
population.loc[population['시군구'] == '세종특별자치시','시군구'] = '세종시'
```

```python
earth['시도+시군구명'] = earth['시도명']+earth['시군구명']
population['시도/시군구명'] = population['도시이름']+ population['시군구']
#시도+시군구명까지 생각해야 되는 것을 알 수 있다.
print(len(earth['시군구명'].unique()),len(earth['시도+시군구명'].unique()))
```

```python
#population이 가지고 있는 시도/시군구명이 earth의 시도+시군구명에 다 있는지 확인 작업. (포항시 북구 / 세종을 찾아냄.)
test1 = list(earth['시도+시군구명'].unique())
```

```python
test2 = list(population['시도/시군구명'].unique())
```

```python
te,te2 = [],[]
for i in test2:
    if i in test1:
        te.append(i)
        
    else:
        te2.append(i)
        
        
#s가 출력이 안되면 시도_시군구명이 다 있다는 것.
for s in test1:
    if s not in te:
        print(s)
```

```python
len(te)   # earth에 있는 (시도+시군구명)이 population에 있는 시도/시군구명안에 다 있는 것을 확인 할 수 있다.
```

```python
te2   #수원시의 (+장안구 +권선구+팔달구+영통구)가 earth에선 통틀어서 수원시로 묶여있다.
#####
#earth를 분화하여 처리 할까??  / 그냥 수원시로 사용할까??(merge로 합치면 상관없긴 하다.)
```

```python
population[population['시군구'].str.contains('수원시')]
```

```python
earth[earth['시군구명'].str.contains('수원시')]
```

```python
#통합
population_total = pd.merge(earth,population[['총인구수','시도/시군구명']],left_on='시도+시군구명', right_on='시도/시군구명')
#전체 row 값이 안변한 것을 볼 수 있다!
```

## 2. population_total 처리

```python
population_total.head()
```

```python
population_total.drop(['시도/시군구명','시도+시군구명'],axis=1, inplace = True)
```

```python
#총 인구수에 ',' 다있는지 확인
population_total[population_total['총인구수'].str.contains(',')]
```

```python
population_total['총인구수'] = population_total['총인구수'].str.replace(',','').astype(int)
```

```python
#이 비율 조절은 어떻게 할 지 생각해봐야 할 듯!
population_total['인구대비 수용가능인원비율'] = population_total['수용가능인원수'] / population_total['총인구수'] *100
```
